```python
def makeBirdDict(filename):
    # This function opens the data file from Thomas et al and extracts the species & family into a dictionary
    #initialize dictionary
    birdDict = {}
    #open the file
    fobj = open(filename, "r", encoding = "ISO-8859-15")
    #read the file & list is stored in "all_records"
    all_records = fobj.readlines()[1:] 
    for line in all_records:
        record_elements = line.split(",") #split lines on comma
        species = record_elements[0]
        family = record_elements[2]
        birdDict[species] = family #stick species and family into birdDict
    return birdDict
makeBirdDict("/home/eeb177-student/Desktop/eeb-177/class-assignments/nature21074-s4.csv")
```

```python
nature = makeBirdDict("/home/eeb177-student/Desktop/eeb-177/class-assignments/nature21074-s4.csv")
species = list(nature.keys())
#sorted(species)
species.sort()
first_species = species[0]
first_fam = nature[first_species]
print( first_species, first_fam)

print("family of " +species[0] + " is " + nature[species[0]])
#species.sort[0]
```
